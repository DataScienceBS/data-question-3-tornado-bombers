```python
import pandas as pd
from bs4 import BeautifulSoup as BS
from urllib import request
import numpy as np

wiki = ('https://en.wikipedia.org/wiki/List_of_deadly_earthquakes_since_1900')
result = request.urlopen(wiki)
soup = BS(result.read(), 'html.parser')
t = soup.find_all(class_='wikitable')[0]
df = pd.read_html(str(t), header=0)[0]

print(df)
```
